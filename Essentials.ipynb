# Import libraries

```python
import pydicom
import nibabel as nib
import scipy.io
import skimage
import matplotlib.pyplot as plt
```

# Reading and Visualizing Medical Images
###### - widely used medical imaging formats

# .dcm

```python
# Load the path 
dicom_file_path = '14.dcm'

# Read the DICOM file
dicom_data = pydicom.dcmread(dicom_file_path)

# Extract pixel data from the DICOM file
dicom_array = dicom_data.pixel_array

# Print metadata and array information
print('Metadata: ', dicom_data)
print('Array: ', dicom_array)
```

```python
# Display the DICOM image using Matplotlib
plt.imshow(dicom_array, cmap=plt.cm.gray)
plt.title('DICOM Image')
plt.axis('off')
plt.show()
```

# .nii

```python
# Load a NIfTI file
nifti_file_path = "BraTS20_Training_003_flair.nii" 
nifti_data = nib.load(nifti_file_path)
nifti_array = nifti_data.get_fdata() # Get array

# Print shape, array, and type information
print('Shape: ',nifti_data.shape)
print('Array: ', nifti_array)
print(type(nifti_data))
```

```python
# Display the NIfTI image
plt.imshow(nifti_array[150], cmap='bone')
plt.axis('off')
plt.show()
```

# .mat

```python
#Read the MAT file
mat_data = scipy.io.loadmat('964_117T1.mat')
print(mat_data)
```

```python
# Print image information or array
mat_image = mat_data['image']
print(mat_image)
```

```python
# Display the image
plt.imshow(mat_image[:, :, 0], cmap='gray')  # Assuming the image is grayscale
plt.title('Mat Image')
plt.axis('off')
plt.show()
```

# .tiff

```python
# Load the TIFF file
tif = skimage.io.imread('ID_0000aee4b.tif')

# Print shape and dtype information
print(tif.shape)
print(tif.dtype)
```

```python
# Display the TIFF image
plt.title('Tiff Image')
plt.imshow(tif, cmap ='gray')
plt.axis('off')
plt.show()
```
